```python
import numpy as np
import pandas as pd
import seaborn as sns
from numpy.random import randn, randint, uniform, sample
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

# Question 1:
Plot using sns.relplot, line plot, box plot on fmri data set


```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri, style='event', size='region')
plt.show()


```

```python
sns.relplot(x='timepoint', y='signal', kind='line', data=fmri, style='event', size='region')
plt.show()
```

```python

sns.catplot(x='timepoint', y='signal', kind='box', data=fmri)
plt.show()
```
